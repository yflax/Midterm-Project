```python

MCI['OCC_DATE'] = pd.to_datetime(MCI['OCC_DATE'])

# Fill empty values in other OCC columns with corresponding parts from OCC_DATE
MCI['OCC_YEAR'] = MCI['OCC_YEAR'].fillna(MCI['OCC_DATE'].dt.year)
MCI['OCC_MONTH'] = MCI['OCC_MONTH'].fillna(MCI['OCC_DATE'].dt.month)
MCI['OCC_DAY'] = MCI['OCC_DAY'].fillna(MCI['OCC_DATE'].dt.day)
MCI['OCC_DOY'] = MCI['OCC_DOY'].fillna(MCI['OCC_DATE'].dt.dayofyear)
MCI['OCC_DOW'] = MCI['OCC_DOW'].fillna(MCI['OCC_DATE'].dt.dayofweek)

# Display the updated DataFrame
print(MCI)
```

```python
#create a list for neighbourhood names
neighbourhood_names = to_demog.columns[6:].tolist()
neighbourhood_names
```

```python
# create list of neighbourhood numbers
row = to_demog.iloc[[0]]

# Initialize the list to store the coordinate values
neigh_nums = []

# Iterate over the neighborhood name columns
for name in neighbourhood_names:
    value = row[name].values[0]  # Get the coordinate value for the specific column
    neigh_nums.append(value)

print(neigh_nums)
```

```python
# CREATE a list for row "Prevalence of low income based on the Low-income measure, after tax (LIM-AT) (%)" for each neigbourhood 
#to find percentage of lowincome population

row = to_demog.iloc[[1131]]

lim_percent = []

#iterate
for name in neighbourhood_names:
    value = row[name].values[0]
    lim_percent.append(value)
    
print(lim_percent)
```

```python
# create a dictionary from both lists

neigh_LIM_dict = dict(zip(neigh_nums,lim_percent))
neigh_LIM_dict
```

```python
# loop thru HOOD_140 to create a list of the corresponding LIM values for each neigbourhood , then add the list as a column to MCI

lowinc_list = []
for value in MCI['HOOD_140']:
    if value in neigh_LIM_dict:
        lowinc_list.append(neigh_LIM_dict[value])
    else:
        lowinc_list.append("unknown")

    
```

```python
lowinc_list
```

```python
# add the lowinc list to the MCI dataframe
MCI['lowinc_percent'] = lowinc_list
```

```python
import sqlite3
from sqlite3 import Error

def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection
```

```python
#load dfs as databses and drop duplicate columns in sql , then save as csv

#Create a connection to the SQLite database
conn = sqlite3.connect('my_database.db')

#Create a cursor object to execute SQL queries
cursor = conn.cursor()

#Upload the 1st half of DataFrame to the 'MCI_17' table in the SQLite database
complete_df.to_sql(name='full_db', con=conn, if_exists='replace', index=False)


#Commit the changes to the database
conn.commit()

alter_table_query = """
    ALTER TABLE full_db DROP COLUMN x;
    ALTER TABLE full_db DROP COLUMN y;
"""
conn.executescript(alter_table_query)


# Read the 'full_db' table from the database into a DataFrame
full_db = pd.read_sql_query("SELECT * FROM full_db", conn)

# Save the DataFrame as a CSV file
full_db.to_csv('full_db.csv', index=False)
```
